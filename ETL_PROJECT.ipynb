LOADIND DATA;

```python
# -----------------------------
# 1. EXTRACT: #lOADING OPEN DATASET;
# -----------------------------
import pandas as pd

url = "https://raw.githubusercontent.com/datasets/exchange-rates/refs/heads/main/data/daily.csv"

df = pd.read_csv(url)
df.head()
```

## TRANSFORMING DATA;





1-Selecting & Renaming Important columns

```python
# Select needed columns
df2 = df[["Date", "Country", "Exchange rate"]].copy()

# Rename for simplicity
df2 = df2.rename(columns={
    "Date": "date",
    "Country": "country",
    "Exchange rate":"Rate"
})

df2.head()
```

2- Filtering Rows

```python
# Picking out chosen countries;
selected_countries = ["United States", "Nigeria", "United Kingdom", "France"]

df_filtered = df2[df2["country"].isin(selected_countries)]

df_filtered.head()
```

```python
#df["Country"].unique()
```

3- Computing New Columns;

```python
df_filtered["inverse_rate"] = 1 / df_filtered["Rate"]
df_filtered.head()
```

LOAD- CLEAN CSV

```python
clean_path = "clean_exchange_rates.csv"

df_filtered.to_csv(clean_path, index=False)

print("File saved:", clean_path)
```
